```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv'
df=pd.read_csv(url)
df.head()
```

```python
#2. High Level Data Understanding:
#a. Find no. of rows & columns in the dataset
#b. Data types of columns.
#c. Info & describe of data in dataframe.
```

```python
#a. Find no. of rows & columns in the dataset
print("\n the shape of array: \n",df.shape)
print("\n")
print("\nnumber of Rows: \n",df.shape[0])
print("\n")
print("\nthe number of Columns: \n",df.shape[1])
```

```python
#b. Data types of columns.
df.dtypes
```

```python
#c. Info & describe of data in dataframe.
df.info
```

```python
df.describe
```

```python
#3. Low Level Data Understanding :

#a. Find count of unique values in location column.
df['location'].nunique
```

```python
#b. Find which continent has maximum frequency using values counts
df['continent'].value_counts()
```

```python
#c. Find maximum & mean value in 'total_cases'.
print("The maximum of Total_cases: \n",df["total_cases"].max())
print("\n")
print("The mean of Total_cases: \n",df["total_cases"].mean())
```

```python
#d. Find 25%,50% & 75% quartile value in 'total_deaths'.
print("Quartile value of 25% of total deaths: \n",df["total_deaths"].quantile(0.25),"\n")

print("Quartile value of 50% of total quartile : \n",df["total_deaths"].quantile(0.5),"\n")

print("Quartile value of 75% of total deaths : \n",df["total_deaths"].quantile(0.75),"\n")
```

```python
#e. Find which continent has maximum 'human_development_index'
df.loc[df["human_development_index"] == df["human_development_index"].max()].head(1)["continent"]
```

```python
#f. Find which continent has minimum 'gdp_per_capita'
df.loc[df["gdp_per_capita"] == df["gdp_per_capita"].min()].head(1)["continent"]
```

```python
#4. Filter the dataframe with only this columns
#['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.
df.filter(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
```

```python
#5. Data Cleaning

#a. Remove all duplicates observations
df.drop_duplicates()
```

```python
#b. Find missing values in all columns
df.isnull()
```

```python
#c. Remove all observations where continent column value is missing
df.dropna(subset=["continent"])
```

```python
#d. Fill all missing values with 0
df.fillna(0)
```

```python
#6. Date time format :

#a. Convert date column in datetime format using pandas.to_datetime
df.dtypes["date"]
```

```python
df["date"] = pd.to_datetime(df['date'])
```

```python
#b. Create new column month after extracting month data from datecolumn.
df["month"] = df["date"].dt.month
df.dtypes
```

```python
df.head(1).month
```

```python
#7. Data Aggregation

#a. Find max value in all columns using groupby function on 'continent'
df1 = df.groupby(["continent"]).max().head(1)
df1
```

```python
# tip after using group_by funvtion
df2 = df1.reset_index()
df2
```

```python
#b. Store the result in a new dataframe named 'df_groupby'.
#(Use df_groupby dataframe for all further analysis)
df_groupby = pd.DataFrame(df)
df_groupby
```

```python
# use df_groupby dataframe for all further analysis
df_groupby.head()
```

```python
#8. Feature Engineering :

#a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df["total_deaths_to_total_cases"] = df["total_deaths"] / df["total_cases"]
df["total_deaths_to_total_cases"]
```

```python
#9. Data Visualization :

#a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
sns.displot(df["gdp_per_capita"],kind = 'hist')
```

```python
sns.displot(df["gdp_per_capita"],kde = False,color = 'yellow',bins = 10)
```

```python
num = df["gdp_per_capita"]
sns.histplot(num,kde = True)
```

```python
#b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
sns.scatterplot(data = df,x = 'total_cases',y = 'gdp_per_capita',hue = 'new_cases')
```

```python
#c. Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby,kind = "hist")
```

```python
#d. Plot a bar plot of 'continent' column with 'total_cases' .
#Tip : using kind='bar' in seaborn catplot

sns.barplot(x = "continent",y = "total_cases",data= df_groupby)
```
